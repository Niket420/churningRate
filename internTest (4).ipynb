```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the data
df  = pd.read_csv('/kaggle/input/dataset/customer_churn_dataset.csv')
df.head()
```

```python
#removing the columns that will produce effect on output

df.drop(['Customer_ID','Gender','Age'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# checking for missing/NULL data
df.isnull().sum()
```

```python
df.describe()
```

```python
#checking for unique value in location
df['Location'].value_counts()
```

```python
df['Product_Preferences'].value_counts()
```

```python
df.info()
```

```python
# visualizing range of each data
sns.boxplot(df)
```

```python
# independent data
x= df.iloc[:,:-1]
x
```

```python
# target data
y= df['Churn']
y
```

```python
# dividing data into train and test

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=33)
```

```python
x_train.head()
```

```python
#converting object data type into numerical value
## to avoid data leakage , encode x_train,x_test separately

from sklearn.preprocessing import OrdinalEncoder



encoder2 = OrdinalEncoder(categories = [['Rural','Urban','Suburban']])
encoder3 = OrdinalEncoder(categories = [['Home Goods','Electronics','Clothing']])



x_train[['Location']] = encoder2.fit_transform(x_train[['Location']])
x_test[['Location']] = encoder2.fit_transform(x_test[['Location']])



x_train[['Product_Preferences']] = encoder3.fit_transform(x_train[['Product_Preferences']])
x_test[['Product_Preferences']] = encoder3.fit_transform(x_test[['Product_Preferences']])



encoder4 = OrdinalEncoder(categories=[['Low','Medium','High']])
x_train[['Discount_Usage']] = encoder4.fit_transform(x_train[['Discount_Usage']])
x_test[['Discount_Usage']] = encoder4.fit_transform(x_test[['Discount_Usage']])


x_train.head()
```

```python
x_train.head(3)
```

```python
# visualizing actual data
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle("Visualizing Data")

custom_labels = ['Location', 'Income', 'Total_Spending', 'Frequency_of_Purchases', 'Loyalty_Program',
                 'Customer_Service_Interactions', 'Time_Since_Last_Purchase', 'Product_Preferences', 'Discount_Usage']

for i in range((x_train.shape[1])):
    plt.subplot(3, 3, i + 1)
    sns.kdeplot(x_train[custom_labels[i]], shade=True, color='r')  # Use column names for indexing
    plt.xlabel(custom_labels[i])

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x_train)
```

```python
# data is not well distributed,therefore scaling is required
```

```python
# scaling the data

from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()


x_train_scaled =scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# visualizing after scaling
```

```python
sns.boxplot(x_train_scaled)
```

```python
plt.figure(figsize=(15,15))
plt.suptitle("visualizing datas")
custom_labels = ['Location','Income','Total_Spending','Frequency_of_Purchases','Loyalty_Program','Customer_Service_Interactions','Time_Since_Last_Purchase','Product_Preferences','Discount_Usage']

for i in range((x_train_scaled.shape[1])):
    plt.subplot(3,3,i+1)
#     sns.histplot(x_train_scaled[:, i], kde=True, color='skyblue', bins=30)
    sns.kdeplot(x=x_train_scaled[i],shade=True,color='r')
    plt.xlabel(custom_labels[i])

plt.tight_layout()
plt.show()
```

```python
# now the data is in fixed range, making it easy to train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
# logistic regression

linearReg = LogisticRegression()
linearReg.fit(x_train_scaled,y_train)
y_pred =linearReg.predict(x_test_scaled)

test_score = accuracy_score(y_test,y_pred)
f1Score = f1_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)


print(test_score)
print(f1Score)
print(recall)
```

```python
# decision tree classifier
```

```python
DTC= DecisionTreeClassifier()
DTC.fit(x_train_scaled,y_train)
y_pred = DTC.predict(x_test_scaled)

test_score = accuracy_score(y_test,y_pred)
f1Score = f1_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

print('test score is',test_score)
print('f1-score is',f1Score)
print('recall is:',recall)
```

```python
## hyperparameter tuning

params = {
    'max_depth': [3, 5, 10,15, None],
    'min_samples_split': [2, 5, 10,20],
    'min_samples_leaf': [1, 2, 4,6],
    'criterion': ['gini', 'entropy']
}



cvDTC = RandomizedSearchCV(
    estimator=DTC,
    param_distributions=params,
    scoring='accuracy',
    cv=5,
    n_iter=10,
    random_state=42,
    verbose=3
)
cvDTC.fit(x_train_scaled,y_train)
y_pred=cvDTC.predict(x_test_scaled)

test_score = accuracy_score(y_test,y_pred)
f1Score = f1_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

print('test score is',test_score)
print('f1-score is',f1Score)
print('recall is:',recall)
```

```python
# random forest classifier
```

```python
randomFor= RandomForestClassifier()
randomFor.fit(x_train_scaled,y_train)
y_pred = randomFor.predict(x_test_scaled)

test_score = accuracy_score(y_test,y_pred)
f1Score = f1_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

print('test score is',test_score)
print('f1-score is',f1Score)
print('recall is:',recall)
```

```python
## hyperparameter tunning for randomForest

params={
    'n_estimators':[10,20,50,80,100],
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_depth':[3,5,10,None]

}

cv = RandomizedSearchCV(estimator=randomFor,param_distributions = params,scoring ='accuracy',cv=5,verbose=3)
cv.fit(x_train_scaled,y_train)
y_pred = cv.predict(x_test_scaled)

test_score = accuracy_score(y_test,y_pred)
f1Score = f1_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

print('test score is',test_score)
print('f1-score is',f1Score)
print('recall is:',recall)
```

```python
# support vector classifier
```

```python
svc= SVC()
svc.fit(x_train_scaled,y_train)
y_pred = svc.predict(x_test_scaled)

test_score = accuracy_score(y_test,y_pred)
f1Score = f1_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

print('test score is',test_score)
print('f1-score is',f1Score)
print('recall is:',recall)
```

```python
### hyperparameter tunning for svm

from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10,50],
              'gamma': [0.01, 0.1, 1, 10],
              'kernel': ['linear']}

# Create a grid search object
cv = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
cv.fit(x_train_scaled,y_train)
y_pred=cv.predict(x_test_scaled)


test_score = accuracy_score(y_test,y_pred)
f1Score = f1_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

print('test score is',test_score)
print('f1-score is',f1Score)
print('recall is:',recall)
```
